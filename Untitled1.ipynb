```python
from PIL import Image

import pyocr
import cv2
```

```python
img1 = Image.open('student.card.jpg')
img1
```

```python
tools = pyocr.get_available_tools()
tools
```

```python
len(tools)
```

```python
tool = tools[0]
tool.get_name()
```

```python
txt1 = tool.image_to_string(
    img1,
    lang = 'jpn',
    builder=pyocr.builders.TextBuilder(tesseract_layout=3)
)
txt1
```
